#Pemilihan Algoritma Klasifikasi

Dari algoritma dan parameternya yang telah dicoba yaitu:
1. Random Forest
2. SVM
3. Decision Tree

Selanjutnya, divisualisasikan hasil metric evaluasinya per masing-masing algoritma supaya dapat diketahui yang terbaik.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Random Forest Evaluation Metrics

eval = ("Accuracy", "Precision", "Recall", "F1-Score")
scheme = {
    "Default":[76.75, 75.69, 78.70, 77.16],

    "Best Manual (Scheme 6)":[76.80, 75.19, 79.87, 77.46],

    "Hyperparameter Tuning":[77.07, 76.38, 78.24, 77.30],
}

x = np.arange(len(eval))  # the label locations
width = 0.3  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained', figsize = (7, 7))

for attribute, measurement in scheme.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute, edgecolor='black')
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('')
ax.set_title('Random Forest Evaluation Metrics')
ax.set_xticks(x + width, eval)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 100)

plt.grid(color = 'green', linestyle = '--', linewidth = 0.25)

plt.show()
```

```python
#SVM Evaluation Metrics

eval = ("Accuracy", "Precision", "Recall", "F1-Score")
scheme = {
    "Default":[74.72, 76.68, 70.98, 73.69],

    "Best Manual (Scheme 9)":[75.27, 76.34, 73.10, 74.68],

    "Hyperparameter Tuning":[76.84, 75.08, 80.23, 77.57],
}

x = np.arange(len(eval))  # the label locations
width = 0.3  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained', figsize = (7, 7))

for attribute, measurement in scheme.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute, edgecolor='black')
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('')
ax.set_title('SVM Evaluation Metrics')
ax.set_xticks(x + width, eval)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 100)

plt.grid(color = 'green', linestyle = '--', linewidth = 0.25)

plt.show()
```

```python
#To Decide Which Algorithm To Use
#Choose 1 from each graph and visualize them in a graph

#Melihat evaluasi metric recall dan F1-Score terbaik, maka di masing-masing algoritma terpilih algoritma terbaik yaitu
#Random Forest = Best Manual (Scheme 6)
#SVM = Hyperparameter Tuning
#Decision Tree = Best Manual (Scheme 3)

eval = ("Accuracy", "Precision", "Recall", "F1-Score")
scheme = {
    "Random Forest":[76.80, 75.19, 79.87, 77.46],

    "SVM":[76.84, 75.08, 80.23, 77.57],

    "Decision Tree":[77.07, 76.06, 78.88, 77.44],
}

x = np.arange(len(eval))  # the label locations
width = 0.3  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in scheme.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute, edgecolor='black')
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('')
ax.set_title('The Best of Classification Algorithm')
ax.set_xticks(x + width, eval)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 100)

plt.grid(color = 'green', linestyle = '--', linewidth = 0.25)

plt.show()
```

#Pemilihan Algoritma Clustering

Dari algoritma dan parameternya yang telah dicoba yaitu:

1. K-Means
2. DBSCAN
3. K-Medoids

Selanjutnya, divisualisasikan hasil metric evaluasinya per masing-masing algoritma supaya dapat diketahui yang terbaik.

```python
#K-Means Evaluation Metrics

eval = ("K-Means (Default", "K-Means (Scheme 8)", "K-Means (Hyperparameter Tuning)",
        "DBSCAN (Default)", "DBSCAN (Scheme 4)", "DBSCAN (Hyperparameter Tuning)",
        "K-Medoids (Default)", "K-Medoids (Scheme 3)", "K-Medoids (Hyperparameter Tuning)"
        )
schemes = (0.61, 0.63, 0.76,
           0.89, 0.79, 0.71,
           0.46, 0.52, 0.48
           )

# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))

# Horizontal Bar Plot
ax.barh(eval, schemes)

# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)

# Add x, y gridlines
ax.grid(color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)

# Show top values
ax.invert_yaxis()

# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')

# Add Plot Title
ax.set_title('Silhouette Coefficient Score',
             loc ='left', )

# Add Text watermark
fig.text(0.9, 0.15, 'Jeeteshgavande30', fontsize = 12,
         color ='grey', ha ='right', va ='bottom',
         alpha = 0.7)

# Show Plot
plt.show()
```

```python
#To Decide Which Algorithm To Use
#Choose 5 which has the highest Silhouette Coefficient Score

#1 = DBSCAN (Default), k = 106
#2 = DBSCAN (Scheme 4), k = 1
#3 = K-Means (Hyperparameter Tuning), k = 9
#4 = DBSCAN (Hyperparameter Tuning), k = 21
#5 = K-Means (Scheme 8), k = 6
```
