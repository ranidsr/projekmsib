```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn_extra.cluster import KMedoids
```

```python
#Data Loading

import pandas as pd

url = "https://github.com/ranidsr/projekmsib/raw/main/E%20Commerce%20Dataset.xlsx"
df = pd.read_excel(url, sheet_name='E Comm')
df
```

```python
#Update Dataframe berdasarkan fitur yang digunakan saja

df=df[['OrderCount', 'CouponUsed', 'HourSpendOnApp']]
df
```

```python
#Lihat rangkuman data apakah ada missing values

df.info()
```

```python
#Handling Missing Values
import numpy as np

df.dropna(axis=0, how='any', subset=(['OrderCount', 'CouponUsed', 'HourSpendOnApp']), inplace=True)
df.reset_index()
df = pd.concat([df], ignore_index=True)
```

```python
df.info()
```

```python
import seaborn as sns

plt.figure(figsize = (18,15))
sns.heatmap(df.corr() , annot = True , cmap = 'Blues')
```

```python
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score, make_scorer, homogeneity_score
from sklearn_extra.cluster import KMedoids
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
X, _ = make_blobs(random_state=42)
X = StandardScaler().fit_transform(X)
X
```

```python
#Hyperparameter Tuning Skema 10
#n_clusters = {3, 4, 5, 6, 7, 8, 9}
#method = {alternate, pam}
#init = {random, heuristic, k-medoids++, build}

kmedoids = KMedoids()

param_grid = {
    'n_clusters': [3, 4, 5, 6, 7, 8, 9],
    'method': ('alternate', 'pam'),
    'init': ('random', 'heuristic', 'k-medoids++', 'build')
}

x = np.random.randn(4013, 2)
y = np.random.randint(0, 3, size=x.shape[0])

grid = GridSearchCV(kmedoids, param_grid=param_grid, cv=5, scoring="rand_score", error_score="raise"
)

grid.fit(x, y)
grid.best_params_
```

```python
#Skema 10
#n_clusters = 9
#method = 'alternate'
#init = 'k-medoids++'


kmedoids10 = KMedoids(n_clusters=9, method='alternate', init='heuristic')
kmedoids10.fit(X)
```

```python
cluster_labels10 = kmedoids10.labels_
```

```python
silhouette_avg10 = silhouette_score(X, cluster_labels10)
```

```python
print(f"Silhouette Score di Skema 10: {silhouette_avg10}")
```

```python
for i in range(9):
  plt.scatter(X[cluster_labels10 == i, 0], X[cluster_labels10 == i, 1], label=f'Cluster {i+1}')

plt.scatter(kmedoids10.cluster_centers_[:,0], kmedoids10.cluster_centers_[:,1], s=200, c='red', label='Medoids')
plt.legend()
plt.title("K-Medoids Clustering")
plt.xlabel("HourSpendOnApp")
plt.ylabel("CouponUsed")
plt.show()
```
